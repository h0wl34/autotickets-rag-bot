### Анализ временных фич

```python
# Приводим даты к типу datetime 
for col in ['OPEN_TIME_', 'RESOLVE_TIME_', 'CLOSE_TIME_', 'ATC_NEXT_BREACH_']:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
bad_resolve_before_open = (df['RESOLVE_TIME_'] < df['OPEN_TIME_']).sum()
bad_close_before_open = (df['CLOSE_TIME_'] < df['OPEN_TIME_']).sum()
bad_resolve_after_close = (df['RESOLVE_TIME_'] > df['CLOSE_TIME_']).sum()

print(f"RESOLVE_TIME_ < OPEN_TIME_: {bad_resolve_before_open}")
print(f"CLOSE_TIME_ < OPEN_TIME_: {bad_close_before_open}")
print(f"RESOLVE_TIME_ > CLOSE_TIME_: {bad_resolve_after_close}")
```

Очевидно, времена переставлены (нас предупреждали о "зашумлении"). Попробуем вернуть всё назад:

```python
mask = df['RESOLVE_TIME_'] < df['OPEN_TIME_']

# для строк, где условие True, меняем местами значения в двух колонках
df.loc[mask, ['OPEN_TIME_', 'RESOLVE_TIME_']] = df.loc[mask, ['RESOLVE_TIME_', 'OPEN_TIME_']].values
```

```python
bad_resolve_before_open = (df['RESOLVE_TIME_'] < df['OPEN_TIME_']).sum()
bad_close_before_open = (df['CLOSE_TIME_'] < df['OPEN_TIME_']).sum()
bad_resolve_after_close = (df['RESOLVE_TIME_'] > df['CLOSE_TIME_']).sum()

print(f"RESOLVE_TIME_ < OPEN_TIME_: {bad_resolve_before_open}")
print(f"CLOSE_TIME_ < OPEN_TIME_: {bad_close_before_open}")
print(f"RESOLVE_TIME_ > CLOSE_TIME_: {bad_resolve_after_close}")
```
